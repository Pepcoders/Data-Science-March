```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
plt.style.use("seaborn")
```

```python
tips.sample(5)
```

```python
bill = tips['total_bill']
print(bill.min())
print(bill.max())
print(bill.shape)
```

```python
plt.figure(figsize= (6, 4))
sns.distplot(tips['total_bill'])
plt.show()
```

```python
plt.figure(figsize= (6, 4))
sns.displot(tips['total_bill'])
plt.show()
```

```python
plt.figure(figsize= (6, 4))
sns.displot(tips['total_bill'], kde = True)
plt.show()
```

```python
plt.figure(figsize= (6, 4))
sns.displot(tips['total_bill'], kde = True, bins = 10)
plt.show()
```

```python
plt.figure(figsize= (6, 4))
sns.displot(tips['total_bill'], bins = 2)
plt.show()
```

```python
data = [2, 3, 4, 5, 1, 2, 6, 7, 9, 10]
sns.displot(data, bins = 3)
plt.show()
```

```python
iris = sns.load_dataset("iris")
```

```python
iris
```

```python
fig = px.scatter(iris, x = "sepal_length", y = "sepal_width", color = "species")
fig.show()
```

```python
fig = px.scatter_3d(iris, x = "sepal_length", y = "sepal_width", z = "petal_length", color = "species")
fig.show()
```

```python
fig = px.bar(iris, x = "sepal_length", y = "sepal_width", color = "species")
fig.show()
```

```python
fig = px.box(iris, x = "species", y = "sepal_width")
fig.show()
```

```python
fig = px.pie(iris, names= "species", values = "sepal_width", color = "species")
fig.show()
```
