```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=500, centers = 4, random_state = 42)

plt.scatter(X[:, 0], X[:, 1], c = y)
plt.show()
```

```python
print(y)
```

```python
X.shape, y.shape
```

```python
from sklearn.cluster import KMeans
km = KMeans() # n_clusters=8

km.fit(X)
```

```python
y_pred = km.predict(X)
y_pred
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y_pred)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], c = 'red', s = 50)
plt.show()
```

```python
km.cluster_centers_
```

```python
km = KMeans(n_clusters=4) # n_clusters=4

km.fit(X)
y_pred = km.predict(X)
plt.scatter(X[:, 0], X[:, 1], c = y_pred)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], c = 'red', s = 100)
plt.show()
```

```python
km.cluster_centers_
```

#Noise Data

```python
X, y = make_blobs(n_samples=500, centers = 4,cluster_std = 3, random_state = 40)

plt.scatter(X[:, 0], X[:, 1], c = y)
plt.show()
```

```python
km = KMeans(n_clusters=4) # n_clusters=4

km.fit(X)
y_pred = km.predict(X)
plt.scatter(X[:, 0], X[:, 1], c = y_pred)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], c = 'red', s = 100)
plt.show()
```

```python
error = []

for k in range(1, 20):
  km = KMeans(n_clusters=k)
  km.fit(X)

  wcss = km.inertia_
  error.append(wcss)
```

```python
plt.figure(figsize = (10, 8))
plt.plot(range(1, 20), error)
plt.show()
```
