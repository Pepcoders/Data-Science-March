```python
import matplotlib.pyplot as plt
```

```python
x = [10, 20, 30, 40, 50, 60]
y = [10, 20, 15, 18, 16, 30]


plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker = 'o')
plt.show()
```

```python
plt.plot(y)
plt.show()
```

```python
help(plt.plot)
```

```python
plt.figure(figsize = (8, 6))
plt.plot(x, y, marker = 'o', ms = 20, mec = 'red', mfc= 'orange')
plt.show()
```

```python
plt.figure(figsize = (8, 6))
plt.plot(x, y, ls = '--', c = "c", lw = 5, marker = 'o', ms = 20, mec = 'red', mfc= 'orange')
plt.show()
```

```python
# marker | line | color

plt.plot(y, '^:b')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker = 'o')
plt.xlabel("Time - >")
plt.ylabel("Temp")
plt.title("polution Temp Graph")
plt.grid()
plt.show()
```

```python
import numpy as np
y = np.arange(10)

plt.plot(y)
plt.plot(y**2)
plt.plot(y**3)
plt.show()
```

```python
plt.figure()
plt.plot(y)
plt.plot(y**2)

plt.figure()
plt.plot(y**3)

plt.show()
```

```python
country = ["Ind", "pak", "Uk", "US", "Russ"]
male = [78, 97, 23, 90, 63, ]
female = [73, 98, 32 ,23, 34]
```

```python
plt.style.use("fivethirtyeight")
plt.bar(country, male)
plt.show()
```

```python
plt.style.use("fivethirtyeight")
plt.bar(country, male, width = 0.5, color="orange")
plt.show()
```

```python
plt.bar(country, female)
plt.bar(country, male, width = 0.5)
plt.show()
```

```python
plt.bar(country, female)
plt.bar(country, male, width = 0.5, align="edge")
plt.show()
```

```python
plt.bar(country, female)
plt.bar(country, male, bottom = female)
plt.show()
```

```python
plt.bar(country, female, label = "Female")
plt.bar(country, male, bottom = female, label = "Male")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend(facecolor = "white", shadow = True, frameon=True)
plt.show()
```
