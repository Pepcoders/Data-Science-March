```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/ushape.csv")
```

```python
df.head()
```

```python
df["class"].value_counts()
```

```python
plt.scatter(df["X"], df["Y"], c = df["class"])
```

```python
X = df[["X", "Y"]].values
y = df["class"].values

X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

plt.scatter(X_train[:, 0], X_train[:, 1], c = y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors= 5)

knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
np.sqrt(X_train.shape[0])
```

```python
X_train.shape
```

```python
k = 7

knn = KNeighborsClassifier(n_neighbors= k)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
accuracy = []

for k in range(1, 16):
  knn = KNeighborsClassifier(n_neighbors= k)
  knn.fit(X_train, y_train)
  y_pred = knn.predict(X_test)

  accuracy.append(accuracy_score(y_test, y_pred))
```

```python
plt.plot(range(1, 16), accuracy)
```

```python
train_error = []
test_error = []

for k in range(1, 16):
  knn = KNeighborsClassifier(n_neighbors= k)
  knn.fit(X_train, y_train)

  train = confusion_matrix(y_train, knn.predict(X_train))
  test = confusion_matrix(y_test, knn.predict(X_test))

  train_error.append((train[0][1] + train[1][0])/ train.sum())
  test_error.append((test[0][1] + test[1][0])/ test.sum())
```

```python
plt.plot(range(1, 16), train_error, label = "Training Error")
plt.plot(range(1, 16), test_error, label = "Testing Error")
plt.legend()
```

```python
knn = KNeighborsClassifier(n_neighbors= 7)
knn.fit(X_train, y_train)


a=np.arange(start=X_train[:,0].min()-1, stop=X_train[:,0].max()+1, step=0.01)
b=np.arange(start=X_train[:,1].min()-1, stop=X_train[:,1].max()+1, step=0.01)


XX,YY=np.meshgrid(a,b)

data = np.array([XX.ravel(),YY.ravel()]).T
labels = knn.predict(data)

plt.contourf(XX,YY,labels.reshape(XX.shape), alpha=0.7)
plt.scatter(X_train[:,0],X_train[:,1], c=y_train)
```

```python
def draw_knn(data_set, k):
  df = pd.read_csv(data_set)

  X = df[["X", "Y"]].values
  y = df["class"].values

  knn = KNeighborsClassifier(n_neighbors = k)
  knn.fit(X, y)
  
  a=np.arange(start=X[:,0].min()-1, stop=X[:,0].max()+1, step=0.01)
  b=np.arange(start=X[:,1].min()-1, stop=X[:,1].max()+1, step=0.01)

  XX,YY=np.meshgrid(a,b)

  data = np.array([XX.ravel(),YY.ravel()]).T
  labels = knn.predict(data)

  plt.contourf(XX,YY,labels.reshape(XX.shape), alpha=0.7)
  plt.scatter(X[:,0],X[:,1], c=y)
```
