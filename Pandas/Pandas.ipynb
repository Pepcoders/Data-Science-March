# Series -> 1D data

# iloc[] vs loc[]

The main distinction between the two methods is:
- `loc` gets rows (and/or columns) with particular labels(Explicitly Indexing).
- `iloc` gets rows (and/or columns) at integer locations(Implicit Index).

```python
A.iloc[0]
```

```python
B.loc['a']
```

```python
C = pd.Series([10, 20, 30, 40], index = [23, 54, 0, 1])
```

```python
C
```

```python
C.iloc[0]
```

```python
C.iloc[1]
```

```python
C.loc[0]
```

```python
C.iloc[0:2]
```

```python
C.loc[0:1]
```

```python
C.loc[54:1]
```

```python
B
```

```python
B['a':'c']
```

## Missing values

```python
tom = {"math":40, 'phy':75, 'chem':46}
jerry = {'math':53, 'phy':73}
bob = {"phy":34, 'chem':23}
```

```python
data = {"Tom": tom, "Jerry":jerry, "Bob":bob}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.info()
```

```python
# copy() use to create a deep copy of data
df1 = df.copy()
df2 = df.copy()
```

```python
df1
```

```python
df2
```

```python
df1["Bob"] = df1["Bob"].fillna(50)
```

```python
df1
```

**Using the inplace=True keyword in a pandas method changes the default behaviour such that the operation on the dataframe doesn't return anything, it instead 'modifies the underlying data' (more on that later). It mutates the actual object which you apply it to**

```python
df1["Jerry"].fillna(30, inplace = True)
```

```python
df1
```

```python
df1["Bob"].fillna(30, inplace = True)
df1
```
